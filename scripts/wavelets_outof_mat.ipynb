```python
import os
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image as img
from scipy.signal import cwt, morlet

```

```python
#  Your directory setup code  
# Adjust the widths variable as needed for your specific application.

print("Folder structure created successfully.")
mat = scipy.io.loadmat('./forces_xyz_raw.mat')

def wavelet_dataset_design(dataset_signals, dataset_labels, axes_of_signal):
    sr = 10000
    area = (11, 10, 506, 379)
    widths = np.arange(1, 128)  # Define the scale for the wavelet transform, modify if needed
    
    for counter, item in enumerate(dataset_signals["baseDatastore"]):
        x = item[3][axes_of_signal_dict[axes_of_signal]]
        cwtmatr = cwt(x, morlet, widths)
        
        plt.cla()
        plt.imshow(np.abs(cwtmatr), aspect='auto', extent=[0, len(x), 1, 128], cmap='jet')
        
        label = dataset_labels["baseDatastore"][counter][2][0]
        folder = label_map[label]
        save_path = f'./data/spec/{axes_of_signal}/{folder}/{str(mat["baseDatastore"][counter][0][0])}'
        
        plt.savefig(save_path, bbox_inches='tight')
        cropped_img = img.open(save_path)
        cropped_img = cropped_img.crop(area)
        cropped_img.save(save_path)
        plt.cla()
    
axes_of_signal_dict = {'x': 0, 'y': 1, 'z': 2}
label_map = {'1': 'sharp', '2': 'used', '3': 'dulled'}

wavelet_dataset_design(mat, mat, 'x')  # For x-axis as an example
```

```python
# create paired visualizations for the given axes_of_signal 
 def evaluate_and_visualize(dataset_signals, dataset_labels, axes_of_signal):
    sr = 10000
    widths = np.arange(1, 128)  # Define the scale for the wavelet transform, modify if needed

    for counter, item in enumerate(dataset_signals["baseDatastore"]):
        x = item[3][axes_of_signal_dict[axes_of_signal]]
        cwtmatr = cwt(x, morlet, widths)

        label = dataset_labels["baseDatastore"][counter][2][0]
        folder = label_map[label]

        # Create a 1x2 subplot: [original signal, wavelet transform]
        fig, axs = plt.subplots(1, 2, figsize=(12, 6))

        # Plot original signal
        axs[0].plot(x)
        axs[0].set_title('Original Signal')
        axs[0].set_xlabel('Time')
        axs[0].set_ylabel('Amplitude')

        # Plot wavelet transform
        axs[1].imshow(np.abs(cwtmatr), aspect='auto', extent=[0, len(x), 1, 128], cmap='jet')
        axs[1].set_title('Wavelet Scalogram')
        axs[1].set_xlabel('Time')
        axs[1].set_ylabel('Scale')
        
        # Save the combined visualization
        save_path = f'./data/spec/{axes_of_signal}/{folder}/{str(mat["baseDatastore"][counter][0][0])}_eval.jpg'
        plt.tight_layout()
        plt.savefig(save_path)
        plt.show()  # Display the plot (optional)

        # Clean up to free memory
        plt.close(fig)

# Call the function
evaluate_and_visualize(mat, mat, 'x')  # For x-axis as an example
```
